```python
import sys
sys.path.append("/scratch/group/csce435-f23/python-3.8.17/lib/python3.8/site-packages")
sys.path.append("/scratch/group/csce435-f23/thicket")
from glob import glob

import matplotlib.pyplot as plt
import pandas as pd

import thicket as th

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
tk = th.Thicket.from_caliperreader(glob("*.cali"))
```

```python
tk.dataframe
```

```python
tk.metadata
```

```python
tk.metadata.sort_values(by=['num_procs', 'InputSize'])
```

```python
tk.statsframe.dataframe["time"] = 1
print(tk.tree())
```

```python
gb1 = tk.groupby(["InputSize", "InputType"])
```

```python
ctk1 = th.Thicket.concat_thickets(
    thickets=list(gb1.values()),
    headers=list(gb1.keys()),
    axis="columns",
    metadata_key="num_procs"
)
ctk1.dataframe
```

```python
num_vals = [65536, 262144, 1048576, 4194304]

for i in num_vals:
    ctk1.dataframe = ctk1.dataframe.reset_index()
    if ("node", "") in ctk1.dataframe.columns:
        ctk1.dataframe = ctk1.dataframe.drop([("node", "")], axis=1)
    ctk1.dataframe = ctk1.dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])
```

```python
ctk1.dataframe
```

```python
threads = sorted(list(set(ctk1.dataframe.index.get_level_values("num_procs"))))
```

```python
idx = pd.IndexSlice

for i in num_vals:
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(
        ctk1.dataframe.loc['main', idx[((i, 'random'),), "Total time"]],
        marker='o',
    )
    ax.set_title(str(i) + " random")
    ax.set_xlabel("Threads")
    ax.set_ylabel("Time (seconds)")
    ax.set_yscale("log", base=10)
    ax.set_xscale("log", base=2)
    ax.grid(True)
    ax.set_xticks(threads)
    ax.legend(["total main time"])
    plt.savefig("mpi_quicksort_graphs/" + str(i) + "_main.png")
    plt.show()
```

```python
idx = pd.IndexSlice

for i in num_vals:
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(
        ctk1.dataframe.loc['comm', idx[((i, 'random'),), "Total time"]],
        marker='o',
        label='comm CPU'
    )
    ax.plot(
        ctk1.dataframe.loc['comp', idx[((i, 'random'),), "Total time"]],
        marker='o',
        label='comp CPU'
    )
    ax.set_title(str(i) + " random")
    ax.set_xlabel("Threads")
    ax.set_ylabel("Time (seconds)")
    ax.set_yscale("log", base=10)
    ax.set_xscale("log", base=2)
    ax.grid(True)
    ax.set_xticks(threads)
    ax.legend()
    plt.savefig("mpi_quicksort_graphs/" + str(i) + "_comm_comp.png")
    plt.show()
```
