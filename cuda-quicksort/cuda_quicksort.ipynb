```python
import sys
sys.path.append("/scratch/group/csce435-f23/python-3.8.17/lib/python3.8/site-packages")
sys.path.append("/scratch/group/csce435-f23/thicket")
from glob import glob

import matplotlib.pyplot as plt
import pandas as pd

import thicket as th

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
tk = th.Thicket.from_caliperreader(glob("*.cali"))
```

```python
tk.dataframe
```

```python
tk.metadata
```

```python
tk.metadata.sort_values(by=['num_threads', 'InputSize'])
```

```python
tk.statsframe.dataframe["time"] = 1
print(tk.tree())
```

```python
gb1 = tk.groupby(["InputSize", "InputType"])
```

```python
ctk1 = th.Thicket.concat_thickets(
    thickets=list(gb1.values()),
    headers=list(gb1.keys()),
    axis="columns",
    metadata_key="num_threads"
)
ctk1.dataframe
```

```python
num_vals = [65536, 262144, 1048576, 4194304]
types = ["1%perturbed", "random", "reverse_sorted", "sorted"]

ctk1.dataframe = ctk1.dataframe.reset_index().drop([("node", "")], axis = 1)
ctk1.dataframe = ctk1.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])

for i in num_vals:
    for j in types:
        ctk1.dataframe = ctk1.dataframe.reset_index().drop(
            [ 
             ((i, j), "nid"), ((i, j), "spot.channel"), 
             ((i, j), "Min time/rank"), ((i, j), "Max time/rank"), ((i, j), "Avg time/rank"), 
             ((i, j), "Min GPU time/rank"), ((i, j), "Max GPU time/rank"), ((i, j), "Avg GPU time/rank"), 
            ], axis=1)
        ctk1.dataframe = ctk1.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])
```

```python
ctk1.dataframe
```

```python
threads = sorted(list(set(ctk1.dataframe.index.get_level_values("num_threads"))))
```

```python
idx = pd.IndexSlice

for i in num_vals:
    fig, ax = plt.subplots(1, 4, figsize=(30, 5))
    iter = 0
    for j in types:
        ax[iter].plot(
            ctk1.dataframe.loc['main', idx[((i, j),), "Total time"]],
            marker='o',
        )
        ax[iter].plot(
            ctk1.dataframe.loc['main', idx[((i, j),), "Total GPU time"]],
            marker='o',
        )
        ax[iter].set_title(str(i) + " " + j)
        ax[iter].set_xlabel("Threads")
        ax[iter].set_ylabel("Time (seconds)")
        ax[iter].set_yscale("log", base=10)
        ax[iter].set_xscale("log", base=2)
        ax[iter].grid(True)
        ax[iter].set_xticks(threads)
        ax[iter].legend(["CPU", "GPU"])
        iter += 1
    plt.savefig("cuda_quicksort_graph/"+str(i)+"_main.png")
    plt.show()
    
```

```python
for i in num_vals:
    fig, ax = plt.subplots(1, 4, figsize=(30, 5))
    iter = 0
    for j in types:
        ax[iter].plot(
            ctk1.dataframe.loc['comm', idx[((i, j),), "Total time"]],
            marker='o',
        )
        ax[iter].plot(
            ctk1.dataframe.loc['comm', idx[((i, j),), "Total GPU time"]],
            marker='o',
        )
        ax[iter].plot(
            ctk1.dataframe.loc['comp', idx[((i, j),), "Total time"]],
            marker='o',
        )
        ax[iter].plot(
            ctk1.dataframe.loc['comp', idx[((i, j),), "Total GPU time"]],
            marker='o',
        )
        ax[iter].set_title(str(i) + " " + j)
        ax[iter].set_xlabel("Threads")
        ax[iter].set_ylabel("Time (seconds)")
        ax[iter].set_yscale("log", base=10)
        ax[iter].set_xscale("log", base=2)
        ax[iter].grid(True)
        ax[iter].set_xticks(threads)
        ax[iter].legend(["comm CPU", "comm GPU", "comp CPU", "comp GPU"])
        iter += 1
    plt.savefig("cuda_quicksort_graph/"+str(i)+"_comm_comp.png")
    plt.show()
```
